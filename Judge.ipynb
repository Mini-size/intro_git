```python
import random

# 手の対応表
HANDS = {
    "g": "グー",
    "c": "チョキ",
    "p": "パー",
}

def cpu_hand():
    """CPUの手をランダムに返す"""
    return random.choice(list(HANDS.keys()))

def judge(player, cpu):
    """
    勝敗判定を返す
    return: "win" / "lose" / "draw"
    """
    if player == cpu:
        return "draw"

    # 勝ち条件
    win_map = {
        "g": "c",  # グーはチョキに勝つ
        "c": "p",  # チョキはパーに勝つ
        "p": "g",  # パーはグーに勝つ
    }

    if win_map[player] == cpu:
        return "win"
    else:
        return "lose"

def input_hand():
    """ユーザー入力を受け取り、正しい手(g/c/p)を返す"""
    while True:
        s = input("手を選んでください [g=グー, c=チョキ, p=パー, q=終了]：").strip().lower()
        if s == "q":
            return "q"
        if s in HANDS:
            return s
        print("入力が不正です。g/c/p/q のどれかを入力してください。")

def main():
    print("=== じゃんけんゲーム ===")
    win = lose = draw = 0

    while True:
        player = input_hand()
        if player == "q":
            break

        cpu = cpu_hand()
        result = judge(player, cpu)

        print(f"あなた：{HANDS[player]}  CPU：{HANDS[cpu]}")

        if result == "win":
            print("結果：あなたの勝ち！")
            win += 1
        elif result == "lose":
            print("結果：あなたの負け…")
            lose += 1
        else:
            print("結果：あいこ")
            draw += 1

        print(f"成績：{win}勝 {lose}敗 {draw}分\n")

    print("\n=== 終了 ===")
    print(f"最終成績：{win}勝 {lose}敗 {draw}分")
    print("また遊んでね！")

if __name__ == "__main__":
    main()
```
